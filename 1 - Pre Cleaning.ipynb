```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn
```

```python
df = pd.read_csv('BDL Feedback 31 May 2020.csv')
df
```

```python
# drop irrelevant columns
df.drop(columns=['Department', 'Job Function', 'Imported Designation','Updated Designation', 'Course Domain', 'CourseID','Rated Date Time (SGT)'],inplace=True)
df.head()
```

```python
## rename columns
df.rename({'Agency': 'agency', 'Agency_short(HR Pulse)': 'agency_short', 'Tag': 'tag', 'Course Name': 'course', 'Rating': 'rating', 'Comment': 'comment'}, axis=1, inplace=True)
df.head()
```

```python
## drop optional modules
df = df[~df.course.str.contains("Optional")]
```

```python
## drop quizzes
df = df[~df.course.str.contains("Quiz")]
```

# Take out specific Tags for shuffling to label

## 1. Communication & Collaboration

```python
from sklearn.utils import shuffle
```

```python
cc = df[df.course.str.contains("BDLCC")]
cc
```

```python
## drop rows with empty comments
cc.dropna(subset=['comment'], inplace=True)
cc
```

```python
# shuffle the data to randomize it
shuffled = shuffle(cc)
shuffled.reset_index(inplace=True, drop=True)
shuffled.head()
```

```python
shuffled.to_csv("shuffled_cc.csv")
```
